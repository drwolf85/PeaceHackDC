```python
import pandas as pd
import matplotlib as plt
import smtplib as smt

%matplotlib inline

def time2digits(date):
    y = date / 10000000000
    M = date / 100000000 - y * 100
    D = date / 1000000 - y * 10000 - M * 100
    h = date / 10000 - y * 1000000 - M * 10000 - D * 100
    m = date / 100 - y * 100000000 - M * 1000000 - D * 10000 - h * 100
    s = date - y * 10000000000 - M * 100000000 - D * 1000000 - h * 10000 - m * 100
    tm = y + M / 12.0 + D / 372.0 + h / 8928.0 + m / 535680.0 + s / 32140800.0
    return tm
```

```python
df = pd.read_csv('result.csv')
newdates = [time2digits(i) for i in df.sort(['DATEADDED'],ascending = [0])['DATEADDED']]
plt.pyplot.plot((x=newdates, y=df['AvgTone'], "o:")
#print df.sort(['DATEADDED'],ascending = [0])[['DATEADDED', 'AvgTone']]
#df.sort(['DATEADDED'],ascending = [0]).plot(x='DATEADDED', y='AvgTone')
```
